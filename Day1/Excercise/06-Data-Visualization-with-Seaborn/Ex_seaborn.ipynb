```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
sns.distplot(tips['total_bill'],kde=False)
#kde -Kernel Density Estimation - sum of kernels or distribution of kernels
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
#same as scatter, color of hex will be darker , where points are dense
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
#regression line
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
#scatter plot
```

```python
sns.pairplot(tips)
#to find relationships btw variables
```

```python
sns.rugplot(tips['total_bill'])
#hash represents data in the bin, so wil be dense when frequency is more
```

```python
#matrix plots
```

```python
flights=sns.load_dataset('flights')
flights
```

```python
#heat map - data should be in matrix form.rows and cols should be same variable
tc=tips.corr()
sns.heatmap(tc,annot=True,cmap='coolwarm')
```

```python
fp=flights.pivot_table(index='month',columns='year',values='passengers')
```

```python
sns.heatmap(fp,cmap='coolwarm',linecolor='white',linewidth=3)
```

```python
#cluster map
sns.clustermap(fp)
#to know how variables are similar to each other
```

```python
sns.clustermap(fp,cmap='coolwarm',standard_scale=1)
#standard_scale is to convert values from 0-1
```

```python
#grids
```

```python
iris=sns.load_dataset('iris')
iris
```

```python
sns.pairplot(iris)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
a=sns.PairGrid(iris)
a.map_diag(sns.histplot,kde=True)
a.map_upper(plt.scatter)
a.map_lower(sns.kdeplot)
#to control what type of plots we want 
```

```python
b=sns.FacetGrid(data=tips,col='time',row='smoker')
b.map(sns.histplot,'total_bill')
```

```python
c=sns.FacetGrid(data=tips,col='time',row='smoker')
c.map(plt.scatter,'total_bill','tip')
#basically to customise plots.
#here rows are smoker variables and columns are time variables.
```

```python
#linear model plot
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
          scatter_kws={'s':100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',row='time',hue='sex'
          ,aspect=0.6)
#aspect ratio -ratio btw height and width
```

```python
#style and color
```

```python
sns.set_style('darkgrid')
#style - darkgrid,whitegrid,ticks
sns.countplot(x='sex',data=tips)
```

```python
sns.set_style('ticks')
#style - darkgrid,whitegrid,ticks
sns.countplot(x='sex',data=tips)
#remove spines
sns.despine()
```

```python
sns.set_style('ticks')
#style - darkgrid,whitegrid,ticks
sns.countplot(x='sex',data=tips)
#remove spines
sns.despine(left=True,bottom=True)
#to remove even the spines in left and bottom
```

```python
#plt.figsize to change figure size
plt.figure(figsize=(12,3))
sns.countplot(x='sex',data=tips)
```

```python
#to print a poster
sns.set_context('poster')
sns.countplot(x='sex',data=tips)
#will highlight the words for better display, default is notebook.
```
