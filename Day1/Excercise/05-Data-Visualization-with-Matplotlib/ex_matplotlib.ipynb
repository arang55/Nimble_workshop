```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=list(map(lambda a: a**2, x))
```

```python
plt.subplot(1,2,1)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.subplot(1,2,2)
plt.plot(y,x)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
